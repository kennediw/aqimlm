# Random Forest Classifier w/ 2020

```python
from matplotlib.ticker import PercentFormatter

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score
```

```python
def create_timeseries_split(beltsville, test_size):
    stop = int(len(beltsville) * (1-test_size));
    
    Train =  beltsville.iloc[0:stop]
    Test = beltsville.iloc[stop+1:-1]

    xtrain = Train.drop(columns="AQI_TRUE")
    ytrain = Train.AQI_TRUE

    xtest = Test.drop(columns="AQI_TRUE")
    ytest = Test.AQI_TRUE
    
    return xtrain, xtest, ytrain, ytest

def create_splits(x, y, rnd_state=0, test_size=0.20):
    return train_test_split(x, y, test_size=test_size, random_state=rnd_state)
```

```python
train_and_test = create_timeseries_split(beltsville, test_size=0.20) 
xtrain =train_and_test[0].to_numpy() 
ytrain = train_and_test[2].to_numpy()
x_train, x_test, y_train, y_test = create_splits(xtrain, ytrain)

print(f'Training sample: [{x_train.shape}, {y_train.shape}], ({type(x_train)}, {type(x_train)})')
print(f'Training Test sample: [{x_test.shape}, {y_test.shape}], ({type(x_test)}, {type(x_test)})')

xtest_ft =train_and_test[1].to_numpy()
ytest_ft = train_and_test[3].to_numpy()

print(f'Test sample: [{xtest_ft.shape}, {ytest_ft.shape}], ({type(xtest_ft)}, {type(xtest_ft)})')
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_pipeline = Pipeline([('scaling', StandardScaler()),
                        ('rf', RandomForestClassifier())
                       ])

param_grid = [{'rf__max_depth': [5,10, 15, 20],
               'rf__n_estimators': [25, 50, 100],
               'rf__class_weight': [None, 'balanced', 'balanced_subsample'],
              }]
```

```python
rf_results = GridSearchCV(estimator=rf_pipeline, param_grid=param_grid, refit='roc_auc_ovo', cv=5, n_jobs=-1)
rf_results = rf_results.fit(x_train, y_train)
rf_yhat = rf_results.predict(x_test)

rf_results.best_estimator_
```

```python
y_testp = rf_results.predict(x_test)
y_testp_rf = rf_results.predict_proba(x_test)


print(f'Validation score: {rf_results.best_score_:.2%}')
print(f'Test score: {rf_results.score(x_test, y_test):.2%}')


from sklearn.metrics import classification_report
print(classification_report(y_test, y_testp))
```

```python
ems = [('rf', rf_results.best_estimator_)]
en = VotingClassifier(estimators=ems, weights=None, voting='soft', n_jobs=-1)
en = en.fit(x_train, y_train)
scores = cross_val_score(estimator=en, X=x_train, y=y_train, cv=10, scoring='roc_auc_ovo')
print(f'ROC AUC OVO {scores.mean():.2f} (+/- {scores.std():.2f}) [Ensemble]')
```

```python
y_testp = en.predict(x_test)
y_testp_en = en.predict_proba(x_test)

print(f'Test score: {en.score(x_test, y_test):.2%}')

from sklearn.metrics import classification_report
print(classification_report(y_test, y_testp))
```

```python
ytestp_ft = en.predict(xtest_ft)
ytestp_en_ft = en.predict_proba(xtest_ft)

print(f'Test score: {en.score(xtest_ft, ytest_ft):.2%}')

from sklearn.metrics import classification_report
print(classification_report(ytest_ft, ytestp_ft))
```

```python
cf_matrix = confusion_matrix(ytest_ft, ytestp_ft)

fig, ax = plt.subplots(figsize=(12,10))

disp = ConfusionMatrixDisplay(confusion_matrix=cf_matrix, 
                              display_labels=rf_results.best_estimator_)
disp.plot(ax=ax)
ax.set_title("Ensemble Confusion Matrix")
figcm = plt.gcf()
plt.show()
plt.draw()
figcm.savefig('cm.png',bbox_inches='tight',dpi=200)
```

```python
x_train,x_test, y_train, y_test = create_splits(train_and_test[0], train_and_test[2])

yTEST = train_and_test[3]
yTEST = pd.concat([yTEST, pd.Series(ytestp_ft, name="ytestp_ft", index=yTEST.index)], axis=1)
yTRAIN = pd.concat([y_test, pd.Series(y_testp, name="y_testp", index=y_test.index)], axis=1)
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15,10))
yTEST.plot(ax=ax1, style="*", alpha=.5, label="Test Data", fontsize=16)
yTRAIN.plot(ax=ax2, style="*", alpha=.5, label="Training Test Data", fontsize=16)
ax1.legend()
ax2.legend()
ax1.set_title("Ensemble Prediction vs Truth Comparison")
ax2.set_title("Ensemble Test vs Truth Comparison")
ax1.set_ylabel("AQI")
ax2.set_ylabel("AQI")
plt.subplots_adjust(hspace=0.4)

figen = plt.gcf()
plt.show()
plt.draw()
figen.savefig('en.png',bbox_inches='tight',dpi=200)
```

Precision: The precision model evaluation metric is the ratio of true positives over true positives plus false positives, i.e. precision = tp / (tp + fp). Precision shows the model's ability not to label a negative sample as positive. Recall: The recall model evaluation metric is the ratio of true positives over true positives plus false negatives, i.e. precision = tp / (tp + fp). Recall shows the model's ability to detect the positive samples. F1 score: The F1 score (or F-beta score, as it's also known) is a weighted harmonic mean of the precision and recal scores, where an F-beta score of 1 is best and 0 is worst. Support The support value shows the number of occurrences of each class in the y_true (or y_test) data.

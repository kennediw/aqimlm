```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mcmillan_mete = ['/Users/SaSa 01/Desktop/mcmillan_data\\mcmillan_data - mete.csv']
mete = pd.read_csv(mcmillan_mete[0])
mete
```

```python
mete = mete.drop(['station'], axis=1)
mete
```

```python
datetime = pd.to_datetime(mete['valid'])
mete['Date'] = datetime
mete
```

```python
mete
```

```python
mete = mete.drop(['valid'], axis=1)
meteor = mete.groupby(by=mete['Date'].dt.date).mean()
meteor
```

```python
# %% Import Packages
import xarray as xr
import glob
from pathlib import Path  

# %% Merge files
# Range of years for data to merge
years = np.arange(2011,2023)

# Path of folder where data is (with * at the end)
path = r'C:/Users/SaSa 01/Desktop/mcmillan_data/pblh/*'

# Create blank arrays to fill with data
pblh = []
time = []

# Loop all years
for year in years:

    # Open file
    files = glob.glob(path+str(year)+'*.nc')
    df = xr.open_dataset(files[0])
    
    # k is the length of data
    k = df.blh.shape[0]   
    [pblh.append(df.blh.data[i][0][0]) for i in np.arange(k)]
    [time.append(df.time.data[i]) for i in np.arange(k)]

# Put data into pandas and name columns
pbl_df = pd.DataFrame({"time": time, "pblh(m)": pblh})

# Find the max for each day ('D') and create value "max_pbl_df"
max_pbl_df = pbl_df.set_index(['time']).resample('D').max()
# Move the new variable out of the index column
max_pbl_df.reset_index(inplace=True)

m=4200
max_pbl_df = max_pbl_df.head(m)

# %% Save to a new csv file

# New file path and name 
new_file = r'C:\Users\SaSa 01\Desktop\mcmillan_data\pblh\era5_boundary_layer_height_dailymax.csv'

# Save to file
filepath = Path()  
filepath.parent.mkdir(parents=True, exist_ok=True)  
max_pbl_df.to_csv(new_file)
```
